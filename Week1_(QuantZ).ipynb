#### Brief Review of Basic Concepts (10 minutes)

**1. A Whimsical Dip into Calculus**

**a. Derivatives: The Financial Speedometers**

Think of derivatives as the financial world's speedometers. Just as a speedometer tells you how fast you're zooming past your less fortunate, pedestrian friends, a financial derivative can tell you how rapidly a stock price is changing. For instance, if you have a stock that’s as volatile as a cat on a hot tin roof, the derivative will show you just how jumpy it is, moment to moment.

**Python Example:**

````python
import sympy as sp

# Define a symbol
x = sp.symbols('x')

# Function representing stock price over time
function = sp.exp(-x**2) + 3  # somewhat resembling market fluctuations

# Calculate the derivative
derivative = sp.diff(function, x)
derivative
````


**b. Integrals: The Financial Accumulators**

Where derivatives can tell you the rate of change, integrals are like the great aggregators of the financial universe. They sum up all the tiny increments of change and give you the whole picture—like calculating the total area under your revenue curve from a lemonade stand over the summer.

**Python Example:**

````python
# Calculate the area under the curve from day 0 to day 10
total_revenue = sp.integrate(function, (x, 0, 10))
total_revenue
````

**1. Partial Derivatives and Their Financial Shenanigans**

Partial derivatives are like the gossip mongers of the financial world, telling you how one variable (like interest rates) might whisper about changes while keeping other variables (like time) constant. They’re crucial for figuring out how a small nudge to one part of your investment could cause ripples across the portfolio.

**2. Gradient Vectors: The Treasure Maps of Optimization**

If you were a pirate (of the legitimate, financial planning sort), gradient vectors would be your treasure maps. They point out the direction in which you should adjust your investments to achieve the highest increase in profit, navigating through the stormy seas of market volatility.

**Python Example:**

````python
from scipy.optimize import minimize

# Function simulating a financial scenario
def objective_function(vars):
    x, y = vars
    return -(x**2 - 4*x + y**2 - y)  # A profit function for two investments

# Find the optimum point
result = minimize(lambda v: -objective_function(v), [0, 0])  # Start at (0,0)
result.x  # The optimum investment points
````

### Optiver Balloon Blowing Test Simulation: Overview (Interesting & Fun)

#### Introduction

The **Optiver Balloon Blowing Test Simulation** is a game designed to evaluate an individual’s decision-making and risk-taking behavior. It challenges players to manage potential rewards by inflating virtual balloons while avoiding the risk of those balloons bursting. The player must balance the decision of whether to "pump" the balloon for greater earnings or "collect" their money before the balloon bursts, which results in losing the earnings from that balloon and incurring a penalty.

This test is commonly used in financial, trading, and decision-making environments, where managing risk under uncertainty is a key skill. The game's simple mechanics simulate real-world scenarios where individuals must weigh risk and reward to make the most profitable decisions.

---

### Game Rules

1. **Pumping the Balloon**:
    - The player can pump the balloon to increase potential earnings. Each pump increases the balloon’s size and adds a fixed amount of money to the current balloon’s potential earnings.
    - With each additional pump, the balloon is closer to reaching its burst point. If the balloon bursts, the player loses all the potential earnings for that balloon and incurs a penalty that reduces their overall wallet balance.
    
2. **Collecting Money**:
    - At any time before the balloon bursts, the player can choose to "collect" the money accumulated from that balloon. When collecting, the player secures the current balloon’s potential earnings and adds them to their total wallet.
    - Once the money is collected, the player moves on to the next balloon, starting with a fresh round.

3. **Bursting Balloons**:
    - The balloon will burst after a certain number of pumps, which is unknown to the player. When the balloon bursts, the player loses all potential earnings from that balloon and incurs a financial penalty to their total earnings.
    - The bursting point varies for each balloon and follows a hidden pattern that the player cannot know in advance.

4. **Penalties for Bursting**:
    - If a balloon bursts, the player not only loses the potential earnings for that balloon but also receives a penalty, which reduces their total wallet balance. The penalty is fixed and applied each time a balloon bursts.

5. **Total Balloons**:
    - The player progresses through a set number of balloons. The total number of balloons is hidden from the player, but there are 30 balloons in total.
    - The goal is to maximize the overall earnings across all balloons while minimizing losses from bursts and penalties.

---

### Game Mechanics

- **Two Buttons**:
    - The game provides the player with two choices during each balloon:
        1. **Pump**: Inflate the balloon and increase potential earnings.
        2. **Collect**: Secure the earnings from the current balloon and move on to the next one.
        
- **Earnings**:
    - Each pump adds a fixed amount of money to the potential earnings for the current balloon.
    - Collecting secures the earnings, which are added to the player's total wallet.

- **Risk and Reward**:
    - The more pumps a player performs, the greater the risk of the balloon bursting. However, with each pump, the potential earnings from that balloon increase.
    - The player must make strategic decisions on how far to push their luck before collecting to maximize total earnings.

---

### Outcomes and What the Test Evaluates

The **Optiver Balloon Blowing Test Simulation** is designed to assess key aspects of an individual's behavior in the context of risk and reward:

1. **Risk Management**:
   - The game tests how well the player can manage risk. Players who can balance the potential rewards of inflating the balloon with the risk of it bursting will perform well. The test evaluates whether the player is too cautious (collecting too early) or too risky (pushing the balloon until it bursts frequently).

2. **Decision-Making Under Uncertainty**:
   - The player is forced to make decisions without full information (i.e., when the balloon will burst). This uncertainty mirrors real-world scenarios where individuals must make decisions with incomplete data, testing their ability to navigate uncertainty.

3. **Strategic Thinking**:
   - The game rewards those who can plan ahead, taking calculated risks to maximize their earnings while minimizing losses. Players who show strategic thinking by timing their pumps and collections well are likely to succeed.

4. **Loss Aversion**:
   - The test also evaluates how a player responds to losses. Frequent balloon bursts may cause frustration or risk-aversion, which can impact how a player approaches subsequent rounds. Players who can remain consistent and avoid emotional decision-making will score better.

5. **Pattern Recognition**:
   - Although the game does not explicitly reveal when each balloon will burst, over time, highly observant players may notice patterns in the bursting behavior. The test rewards those who can use subtle cues to improve their strategy.

---

### Scoring and Performance Evaluation

At the end of the game, the player is given a **performance score** between 1 and 10 based on their overall results. This score is calculated by considering the following factors:

1. **Total Earnings**:
   - The amount of money the player accumulated across all balloons by collecting before the balloon burst.
   
2. **Penalties**:
   - The number of penalties the player incurred from balloon bursts. Penalties reduce the total earnings and impact the final score.

3. **Optimal Decision-Making**:
   - The score reflects how well the player balanced risk and reward across all balloons. A higher score indicates that the player made strategic decisions, minimized penalties, and maximized total earnings.

---

### Conclusion

The **Optiver Balloon Blowing Test Simulation** is an engaging tool for assessing risk tolerance, decision-making, and strategic thinking under uncertainty. It challenges players to manage their greed for higher rewards while tempering their actions with caution to avoid losses. By evaluating how players handle this balance, the game provides insights into their ability to perform in high-stakes, risk-filled environments—skills that are critical in trading, finance, and decision-making roles.

The simulation's simplicity belies its complexity, and it offers a valuable window into a player's psychological approach to risk, reward, and loss aversion.

````python
# @title Optiver Balloon Blowing Test Simulation {"display-mode":"form"}
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import Button, VBox, HBox, Output, Label, IntProgress, Layout
from IPython.display import display, clear_output

class OptiverBalloonGame:
    def __init__(self):
        self.output = Output()
        self.total_earnings = 0.00
        self.total_penalty = 0.00
        self.balloon_count = 0
        self.max_balloons = 30  # Hidden from the player
        self.game_over = False
        self.prime_sequence = self.generate_prime_sequence(self.max_balloons)
        self.init_new_balloon()
        self.create_widgets()
        self.update_display()

    def generate_prime_sequence(self, n):
        # Generate the first n prime numbers
        primes = []
        num = 2
        while len(primes) < n:
            is_prime = all(num % i != 0 for i in range(2, int(np.sqrt(num)) + 1))
            if is_prime:
                primes.append(num)
            num += 1
        return primes

    def init_new_balloon(self):
        self.current_earnings = 0.00
        self.balloon_popped = False
        self.balloon_size = 0
        if self.balloon_count < self.max_balloons:
            self.explosion_point = self.prime_sequence[self.balloon_count]  # Use prime numbers as explosion points
        else:
            self.explosion_point = np.random.randint(5, 15)  # Fallback explosion point
        self.pump_history = []

    def create_widgets(self):
        self.pump_button = Button(description='Pump', layout=Layout(width='100px'))
        self.collect_button = Button(description='Collect Money', layout=Layout(width='100px'))
        self.pump_button.on_click(self.pump_balloon)
        self.collect_button.on_click(self.collect_earnings)
        self.total_earnings_label = Label(value=f"Total Earnings: ${self.total_earnings:.2f}")
        self.current_earnings_label = Label(value=f"Current Earnings: ${self.current_earnings:.2f}")
        self.status_label = Label(value="Status: Ready to play!")
        self.progress_bar = IntProgress(value=0, min=0, max=113, description='Balloon Size:')
        self.ui = VBox([
            HBox([self.pump_button, self.collect_button]),
            self.total_earnings_label,
            self.current_earnings_label,
            self.status_label,
            self.progress_bar,
            self.output
        ])
        display(self.ui)

    def pump_balloon(self, b):
        if self.game_over:
            self.status_label.value = "Status: Game over. Thank you for playing!"
            return
        if not self.balloon_popped:
            self.balloon_size += 1
            self.current_earnings += 0.10  # Each pump increases potential earnings by $0.10
            self.pump_history.append(1)
            if self.balloon_size >= self.explosion_point:
                self.balloon_popped = True
                self.current_earnings = 0.00  # Lose earnings for this balloon
                penalty = 0.50  # Penalty for popping a balloon
                self.total_penalty += penalty
                self.total_earnings -= penalty  # Apply penalty to total earnings
                self.status_label.value = f"Status: The balloon popped! Penalty applied: ${penalty:.2f}"
                self.check_game_over()
            else:
                self.status_label.value = f"Status: Pumped! Potential earnings: ${self.current_earnings:.2f}"
        else:
            self.status_label.value = "Status: Balloon popped. Please proceed."
        self.update_display()

    def collect_earnings(self, b):
        if self.game_over:
            self.status_label.value = "Status: Game over. Thank you for playing!"
            return
        if not self.balloon_popped:
            self.total_earnings += self.current_earnings
            self.status_label.value = f"Status: Collected ${self.current_earnings:.2f} from this balloon."
            self.current_earnings = 0.00
            self.next_balloon()
        else:
            self.status_label.value = "Status: Balloon popped. Please proceed."
        self.update_display()

    def next_balloon(self):
        self.balloon_count += 1
        if self.balloon_count >= self.max_balloons:
            self.game_over = True
            self.show_final_score()
        else:
            self.init_new_balloon()

    def check_game_over(self):
        self.next_balloon()

    def update_display(self):
        self.total_earnings_label.value = f"Total Earnings: ${self.total_earnings:.2f}"
        self.current_earnings_label.value = f"Current Earnings: ${self.current_earnings:.2f}"
        self.progress_bar.value = self.balloon_size
        with self.output:
            clear_output(wait=True)
            self.draw_balloon()

    def draw_balloon(self):
        plt.figure(figsize=(4,6))
        plt.title("Balloon Simulation")
        circle = plt.Circle((0.5, 0.5), 0.1 + 0.005*self.balloon_size, color='salmon', ec='black')
        ax = plt.gca()
        ax.add_patch(circle)
        plt.xlim(0,1)
        plt.ylim(0,1.5)
        plt.axis('off')
        plt.show()

    def show_final_score(self):
        score = self.calculate_score()
        self.status_label.value = f"Game Over! Your performance score: {score}/10"
        self.pump_button.disabled = True
        self.collect_button.disabled = True

    def calculate_score(self):
        # Benchmarking the user's performance
        max_possible_earnings = sum([(ep - 1) * 0.10 for ep in self.prime_sequence])  # Max earnings if collecting just before bursting
        earnings_after_penalty = max(self.total_earnings, 0)
        performance_ratio = earnings_after_penalty / max_possible_earnings
        score = int(performance_ratio * 10)
        return min(max(score, 1), 10)  # Ensure score is between 1 and 10

# Start the game
game = OptiverBalloonGame()
````
